```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ppd_out = np.fromfile(open("./ppd_capture.mat"), dtype=np.complex64)
flag_detect_out = np.fromfile(open("./flag_detector_capture.mat"), dtype=np.complex64)
fsamp = 400e3
Tsamp = 1 / fsamp
datarate = 50e3

time_ppd = np.linspace(
    start=0, stop=len(ppd_out) * Tsamp, num=len(ppd_out), endpoint=True
)
time_flag = np.linspace(
    start=0, stop=len(flag_detect_out) * Tsamp, num=len(flag_detect_out), endpoint=True
)

fig, ax = plt.subplots(2, 1, figsize=(20, 6))

ax[0].plot(time_ppd, np.abs(ppd_out))
ax[1].plot(time_flag, np.abs(flag_detect_out))
ax[0].set_ylabel("Magnitude of signal")
ax[1].set_ylabel("Magnitude of signal")
ax[1].set_xlabel("Time [s]")
ax[0].set_ylim([0, 0.3])
ax[1].set_ylim([0, 0.3])
ax[0].set_title("Output of hardware accelerated PPD")
ax[1].set_title("Output of flag detector")
```
