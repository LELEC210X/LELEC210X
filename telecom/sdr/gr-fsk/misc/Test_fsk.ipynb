```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
```

```python
fsk_out = np.fromfile(Path(__file__).parent / "./fsk_capture.mat", dtype=np.complex64)
fsamp = 400e3
Tsamp = 1 / fsamp
datarate = 50e3

time_ppd = np.linspace(
    start=0, stop=len(fsk_out) * Tsamp, num=len(fsk_out), endpoint=True
)
fig, ax = plt.subplots(2, 1, figsize=(20, 6))

ax[0].plot(time_ppd, np.real(fsk_out))
ax[1].plot(time_ppd, np.imag(fsk_out))
ax[0].set_ylabel("Magnitude of signal")
ax[1].set_ylabel("Magnitude of signal")
ax[1].set_xlabel("Time [s]")
ax[0].set_ylim([0, 0.3])
ax[1].set_ylim([0, 0.3])
ax[0].set_title("Output of hardware accelerated PPD")
ax[1].set_title("Output of flag detector")
```
